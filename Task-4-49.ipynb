Два насоса заполняют два одинаковых бака водой. Вместе они заполнили бы один бак за з 3 часа. За какое
время каждый из них заполняет свой бак, если известно, что первый насос делает это на 2,5 часа быстрее
второго?

```python
from sympy import Eq
from sympy import solve
from sympy import symbols

x = symbols("x")
eq1 = Eq(1 / x + 1 / (x + 2.5), 1/3)
sol = solve(eq1,x)
print(sol)
print([sol[1], sol[1] + 2.5])
```

```python
from sympy import Eq
from sympy import solve
from sympy import symbols

x = symbols("x")
eq1 = Eq(1 / (1 / x + 1 / (x + 2.5)), 3)
sol = solve(eq1,x)
print(sol)
print([sol[1], sol[1] + 2.5])
```

```python

from sympy import Eq
from sympy import solve
from sympy import symbols

x, y = symbols("x y")
eq1 = Eq(y - x, 2.5)
eq2 = Eq(1 / x + 1 / y, 1/3)
sol = solve([eq1, eq2], [x, y])
print(sol)
print(sol[1])
```

```python
from sympy import Eq
from sympy import solve
from sympy import symbols

x, y = symbols("x y")
eq1 = Eq(y - x, 2.5)
eq2 = Eq(1 / (1 / x + 1 / y), 3)
sol = solve([eq1, eq2], [x, y])
print(sol)
print(sol[1])
```
